For this project I will do an exploratory analysis of the following variables per community area:
* per capita income
* hardship_index 
* crime count
* safety_score 
* rate_of_misconduct_per_100_students
* college_enrollment

I will extract all these variables and join them in a final table by the community area number which is a common column in all tables.

```python
#Crime table has no community_area_name column so we will use the com. area number to extract the com. area name.
#Use the socioeconomic table to extract the comm. area name for each row
%sql select socioeconomic.community_area_name, count(crime.id) as crime_count from crime JOIN socioeconomic \
ON socioeconomic.community_area_number = crime.community_area_number \
group by socioeconomic.community_area_name
```

```python
#pass the above SQL result set into a Python object
table1= %sql select socioeconomic.community_area_name, count(crime.id) as crime_count from crime JOIN socioeconomic \
ON socioeconomic.community_area_number = crime.community_area_number \
group by socioeconomic.community_area_name
```

```python
#pass the above SQL result set into a Python DataFrame
import pandas as pd
df_crime= pd.DataFrame(table1)
df_crime.columns=['community_area_name','crime_count']
```

```python
%sql select socioeconomic.community_area_name, socioeconomic.per_capita_income, socioeconomic.hardship_index from socioeconomic
```

```python
#Order the socioeconomic table by comm. area name in preparation to merge with the other tables
#Delete the 'CHICAGO' row since the comm. area name is missing for those entries
#pass the result set to a Python object
table2= %sql select socioeconomic.community_area_name, CAST(avg(socioeconomic.per_capita_income) AS DECIMAL(10,0)) as per_capita_income, CAST(avg(socioeconomic.hardship_index) AS DECIMAL(10,0)) as hardship_index from socioeconomic where community_area_name!='CHICAGO' group by community_area_name
```

```python
#pass the above SQL result set into a Python DataFrame
df_socioeconomic= pd.DataFrame(table2)
df_socioeconomic.columns=['community_area_name','per_capita_income','hardship_index']
```

```python
#SQL query for the AVG safety_score, SUM college_enrollment, and AVG misconducts_per_100 grouped by community area

table3= %sql select socioeconomic.community_area_name, CAST(avg(safety_score) AS DECIMAL(10,2)) as safety_score, \
sum(college_enrollment) as college_enrollment, CAST(avg("Rate_of_Misconducts__per_100_students_") AS DECIMAL(10,2)) \
as misconducts_per_100 from school, socioeconomic \
where socioeconomic.community_area_number=school.community_area_number \
group by socioeconomic.community_area_name
```

```python
#Pass the SQL resultset above into a Python dataframe
df_school= pd.DataFrame(table3)
df_school.columns=['community_area_name','safety_score','college_enrollment','misconducts_per_100']
```

```python
#Merge the first two dataframes on community_area_name
result1= pd.merge(df_socioeconomic, df_crime, how="left", on="community_area_name")
```

```python
#Merge with the third dataframe on community_area_name
result= pd.merge(result1, df_school, how='left', on="community_area_name")
```

```python
#Print the head of the final dataframe
result.head()
```

```python
#Since all our measures are numerical I will be doing a correlaion matrix
#Create new dataframe including all numerical columns to be included in the correlation matrix
numerical_cols=[]
r=len(list(result))
for i in range(1,r):
    numerical_cols.append(result.iloc[:,i])
```

```python
matrix= pd.DataFrame(numerical_cols).T #transposed dataframe
matrix = matrix.astype(float)
```

```python
#Correlation matrix
matrix.corr()
```

### Correlations:
    hardship_index | per_capita_income: strong negative correlation (-0.85)
    hardship_index | safety_score: strong negative correlation (-0.72)
    safety_score | per_capita_income: strong positive correlation (0.7)
    safety_score | misconducts_per_100: moderate to strong negative correlation (-0.63)

```python
pd.plotting.scatter_matrix(matrix, figsize=(15,15));
```

### Observations: 
    As the per capita income of the area increases the safety score increases.
    The hardship index is negative correlated to both per capita income and safety score.
    The rate of school misconducts is only correlated to the area's safety score.
    The crime count is not correlated to the safety score, the income or the hardship index of the area.
    The crime count is moderately correlated to the rate of college enrollment.
